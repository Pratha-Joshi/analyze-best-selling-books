**Loading the Data**

```python
import pandas as pd
df = pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

**Exploring the Data**

```python
# Get the first 5 rows of the spreadsheet
print(df.head())

# Get the shape of the spreadsheet
print(df.shape)

# Get the column names of the spreadsheet
print(df.columns)

# Get summary statistics for each column
print(df.describe())
```

**Cleaning the Data**

```python
df.drop_duplicates(inplace=True)
```

**Reanming some of the columns**

```python
df.rename(columns={"Name": "Title", "Year": "Publication Year", "User Rating": "Rating","Reviews":"Number of Reviews"}, inplace=True)
```

```python
df.head()
```

**Converting data type of "Price"**

```python
df["Price"] = df["Price"].astype(float)
```

**Analysing the Data**

```python
#Finding how many books of an author have appeared in the bestsellers
author_counts = df['Author'].value_counts()
print(author_counts)
```

```python
#Finding average rating of each genre in the dataset
avg_rating_by_genre = df.groupby('Genre')['Rating'].mean()
print(avg_rating_by_genre)
```

```python
#Finding average price of each genre in the dataset
avg_price_by_genre = df.groupby('Genre')['Price'].mean()
print(avg_price_by_genre)
```

```python
#Finding average reviews for each genre in the dataset
avg_reviews_by_genre = df.groupby('Genre')['Number of Reviews'].mean()
print(avg_reviews_by_genre)
```

**Exporting the Results**

```python
# Export top selling authors to a CSV file
author_counts.head(10).to_csv("top_authors.csv")

# Export average rating by genre to a CSV file
avg_rating_by_genre.to_csv("avg_rating_by_genre.csv")

# Export average price by genre to a CSV file
avg_price_by_genre.to_csv("avg_price_by_genre.csv")

# Export average reviews by genre to a CSV file
avg_reviews_by_genre.to_csv("avg_reviews_by_genre.csv")
```
